# Plot serial intervals

```python
# Import libraries
import numpy as np
from matplotlib import pyplot as plt
import os
import pandas as pd
```

```python
# Read serial interval
si_file = 'si-epinow'
path = os.path.join('../data_library/serial_interval', '{}.csv'.format(si_file))
serial_interval = pd.read_csv(path, header=None)
serial_interval = serial_interval.fillna(0)
serial_intervals = serial_interval.values.T
serial_intervals = np.insert(serial_intervals, 0, np.zeros(np.shape(serial_intervals)[0]), axis=1)
```

```python
plt.figure(figsize=(7.5, 6))
for i in range(np.shape(serial_intervals)[0]):
    plt.plot(serial_intervals[i,:25], 'k', linewidth= 1, alpha=0.1)
plt.xlabel('Time (days)')
plt.ylabel('Probability')
plt.savefig('serial_interval.pdf')
```
